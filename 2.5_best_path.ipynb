```python
list_1 = [chr(x+97).upper() for x in ind[0]] # these are the nodes
```

```python
list_2 = [chr(x+97).upper() for x in ind[1]] # these can be treated as the edges
```

```python
# generate the graph
```

```python
graph = {}
# go over all the position and key
for pos,key in enumerate(list_1):
    # whether key is in the graph
    if key in graph:
        graph[key].append(list_2[pos])
    else:
        graph[key] = [list_2[pos]]
```

```python
graph
```

# 2 Find all the possible path

```python
def findAllPath(graph,start,end,path=[]):
    path = path +[start]
    if start == end:
        return [path]
 
    paths = [] #store all paths    
    for node in graph[start]:
        if node not in path:
            newpaths = findAllPath(graph,node,end,path) 
            for newpath in newpaths:
                paths.append(newpath)
    return paths
```

```python
allpath = findAllPath(graph,'D','K')
print('\nall pathï¼š',allpath)
```

# 3 Build the transition matrix

```python
keylist = [chr(i) for i in range(97, 112)] # match the index and columns with nodes in the graph
```

```python
keylist = [item.upper() for item in keylist]
```

```python
connection_dataframe=pd.DataFrame(connection_matrix.A,index= keylist,columns = keylist)
```

```python
connection_dataframe
```

```python
row_sum = connection_dataframe.sum(axis = 1) # this is a symmetric matrix, so row_sum equals the column_sum
```

```python
propobility_matrix = connection_dataframe.div(row_sum, axis='rows')
```

```python
propobility_matrix
```

```python
propobility_matrix.sum(axis = 1)
```

# 4 Calculate the probability of every path

```python
result = [ ]
for j in range(len(allpath)):
    times = 1
    for i in range(len(allpath[j])-1):       
        times = times * propobility_matrix[allpath[j][i]][allpath[j][i+1]]
    result.append(times)
# print(result)  
```

```python
len(result)
```

```python
result_sort = sorted(result,reverse = True)
```

```python
print('most possible path is : {}'.format(allpath[result.index(result_sort[0])]))
```
